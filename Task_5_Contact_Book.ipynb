```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize contact list
contacts = []

# Function to update contact list display
def update_contact_list():
    contact_list.options = [f"{idx + 1}. {contact['name']} - {contact['phone']}" for idx, contact in enumerate(contacts)]

# Function to add a new contact
def add_contact(_):
    name = name_input.value
    phone = phone_input.value
    email = email_input.value
    address = address_input.value

    if not name or not phone or not email or not address:
        output_area.value = "All fields are required!"
        return

    contacts.append({"name": name, "phone": phone, "email": email, "address": address})
    name_input.value = phone_input.value = email_input.value = address_input.value = ""
    update_contact_list()
    output_area.value = f"Contact '{name}' added successfully!"

# Function to view selected contact
def view_contact(_):
    if not contact_list.value:
        output_area.value = "Please select a contact to view!"
        return

    index = int(contact_list.value.split(".")[0]) - 1
    contact = contacts[index]
    details = (
        f"**Name**: {contact['name']}\n"
        f"**Phone**: {contact['phone']}\n"
        f"**Email**: {contact['email']}\n"
        f"**Address**: {contact['address']}"
    )
    output_area.value = details

# Function to search for a contact
def search_contact(_):
    query = search_input.value
    if not query:
        output_area.value = "Enter a name or phone number to search!"
        return

    results = [f"{idx + 1}. {contact['name']} - {contact['phone']}" for idx, contact in enumerate(contacts) if query in contact["name"] or query in contact["phone"]]
    if not results:
        output_area.value = "No matching contacts found."
    else:
        output_area.value = "\n".join(results)

# Function to delete a contact
def delete_contact(_):
    if not contact_list.value:
        output_area.value = "Please select a contact to delete!"
        return

    index = int(contact_list.value.split(".")[0]) - 1
    contact = contacts[index]
    contacts.pop(index)
    update_contact_list()
    output_area.value = f"Contact '{contact['name']}' deleted successfully!"

# Input widgets
name_input = widgets.Text(placeholder="Enter Name")
phone_input = widgets.Text(placeholder="Enter Phone Number")
email_input = widgets.Text(placeholder="Enter Email")
address_input = widgets.Text(placeholder="Enter Address")
search_input = widgets.Text(placeholder="Enter Name or Phone Number to Search")

# Buttons
add_button = widgets.Button(description="Add Contact", button_style="success")
view_button = widgets.Button(description="View Contact", button_style="info")
search_button = widgets.Button(description="Search Contact", button_style="warning")
delete_button = widgets.Button(description="Delete Contact", button_style="danger")

# Contact list display
contact_list = widgets.Select(options=[], rows=8, description="Contacts:")

# Output area for messages
output_area = widgets.Textarea(value="", placeholder="Messages will appear here.", rows=5, disabled=True)

# Bind button clicks to functions
add_button.on_click(add_contact)
view_button.on_click(view_contact)
search_button.on_click(search_contact)
delete_button.on_click(delete_contact)

# Layout
display(
    widgets.VBox([
        widgets.HTML("<h1>Contact Book</h1>"),
        widgets.HTML("<b>Add Contact</b>"),
        name_input,
        phone_input,
        email_input,
        address_input,
        add_button,
        widgets.HTML("<hr>"),
        widgets.HTML("<b>Contact List</b>"),
        contact_list,
        view_button,
        delete_button,
        widgets.HTML("<hr>"),
        widgets.HTML("<b>Search Contact</b>"),
        search_input,
        search_button,
        widgets.HTML("<hr>"),
        widgets.HTML("<b>Output</b>"),
        output_area
    ])
)
```
