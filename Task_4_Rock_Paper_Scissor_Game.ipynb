```python
import random

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "scissors" and computer_choice == "paper") or
        (user_choice == "paper" and computer_choice == "rock")
    ):
        return "user"
    else:
        return "computer"

def display_result(user_choice, computer_choice, winner):
    print(f"\nYou chose: {user_choice}")
    print(f"Computer chose: {computer_choice}")
    if winner == "tie":
        print("It's a tie!")
    elif winner == "user":
        print("You win!")
    else:
        print("You lose!")

def play_game():
    user_score = 0
    computer_score = 0

    print("Welcome to Rock-Paper-Scissors!")
    print("Instructions: Type 'rock', 'paper', or 'scissors' to play. Type 'exit' to quit the game.\n")

    while True:
        user_choice = input("Enter your choice (rock, paper, scissors): ").lower()

        if user_choice == "exit":
            print("\nThank you for playing!")
            print(f"Final Score: You {user_score} - {computer_score} Computer")
            break

        if user_choice not in ["rock", "paper", "scissors"]:
            print("Invalid choice! Please choose rock, paper, or scissors.\n")
            continue

        computer_choice = get_computer_choice()
        winner = determine_winner(user_choice, computer_choice)

        if winner == "user":
            user_score += 1
        elif winner == "computer":
            computer_score += 1

        display_result(user_choice, computer_choice, winner)
        print(f"Score: You {user_score} - {computer_score} Computer\n")

# Start the game
play_game()
```

# **USER INTERFACE**

```python
import random
from IPython.display import display, clear_output
import ipywidgets as widgets

# Game logic functions
def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "scissors" and computer_choice == "paper") or
        (user_choice == "paper" and computer_choice == "rock")
    ):
        return "user"
    else:
        return "computer"

# Callback for game logic
def play_game(user_choice):
    global user_score, computer_score

    # Get computer choice
    computer_choice = get_computer_choice()
    winner = determine_winner(user_choice, computer_choice)

    # Update scores
    if winner == "user":
        global user_score
        user_score += 1
        result_text.value = f"You chose {user_choice}. Computer chose {computer_choice}. **You win this round!** 🎉"
    elif winner == "computer":
        global computer_score
        computer_score += 1
        result_text.value = f"You chose {user_choice}. Computer chose {computer_choice}. **Computer wins this round!** 😔"
    else:
        result_text.value = f"You chose {user_choice}. Computer chose {computer_choice}. **It's a tie!** 😐"

    # Update score display
    score_text.value = f"Score: **You {user_score} - {computer_score} Computer**"

# Reset game callback
def reset_game(b):
    global user_score, computer_score
    user_score = 0
    computer_score = 0
    result_text.value = "Game has been reset. Make your move!"
    score_text.value = "Score: **You 0 - 0 Computer**"

# Initialize scores
user_score = 0
computer_score = 0

# Widgets for user interaction
rock_button = widgets.Button(description="Rock", button_style="primary")
paper_button = widgets.Button(description="Paper", button_style="info")
scissors_button = widgets.Button(description="Scissors", button_style="success")
reset_button = widgets.Button(description="Reset Game", button_style="warning")

result_text = widgets.HTML(value="Make your move!")
score_text = widgets.HTML(value="Score: **You 0 - 0 Computer**")

# Assign button callbacks
rock_button.on_click(lambda b: play_game("rock"))
paper_button.on_click(lambda b: play_game("paper"))
scissors_button.on_click(lambda b: play_game("scissors"))
reset_button.on_click(reset_game)

# Display the UI
display(widgets.VBox([
    widgets.HTML("<h1>Rock-Paper-Scissors Game</h1>"),
    widgets.HBox([rock_button, paper_button, scissors_button]),
    result_text,
    score_text,
    reset_button
]))
```
